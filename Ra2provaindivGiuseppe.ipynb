# Questão 3 - Vale até 4 Pontos

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


##A)

```python


import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):

  return math.e**(2*x) - math.e**x

def df(x):

  return 2*math.e**(2*x) - math.e**x

def d2f(x):

  return 4*math.e**(2*x) - math.e**x


x = np.linspace(-2, 0.5, 400)


y = f(x)
dy = df(x)
d2y = d2f(x)


plt.plot(x, y, label="f(x)")
plt.plot(x, dy, label="f'(x)")
plt.plot(x, d2y, label="f''(x)")
plt.grid()
plt.ylim(-1,10)
plt.legend()
plt.show()
```

# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


Seja $A$ a área total da lata (em unidades quadradas), $r$ o raio da base da lata (em unidades) e $h$ a altura da lata (em unidades).

Substituindo $2\pi r$ por $d$ (diâmetro), temos:

$$A = dhr + 8r^2$$

Isolando $h$, temos:

$$h = \frac{A - 8r^2}{dr}$$

Encontrando a derivada de $h$ em relação a $d$:

$$\frac{dh}{dd} = \frac{(-16r * dr - A + 8r^2)}{dr^2}$$

Igualando a derivada a zero e resolvendo para $\frac{h}{d}$, temos:

$$\frac{h}{d} = -8$$

**Conclusão:**

A proporção entre a altura e o diâmetro para a lata mais econômica é de $-8$. Isso significa que, para cada unidade de diâmetro, a altura ideal é de 8 unidades.

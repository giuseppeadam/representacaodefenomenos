## Questão 1

Considerando um fenômeno que possa ser explicado pela equação

$$y = 3x^3-150 sin(x)^2+2$$

Encontre o coeficiente angular da tangente a dois pontos que você ache que sejam relevantes para o entendimento deste fenômeno. Você deverá fazer a análise algébrica do processo de determinação do coeficiente angular e apresentar todas a passagens matemáticas, e justificativas que o levaram a escolha destes pontos e o que você aprendeu sobre este fenômeno. Além disso, toda a álgebra deve estar explícita em notação matemática (use latex) em um bloco de textos deste caderno.

Valor da Questão: até 1 ponto pela apresentação algébrica dos coeficientes angulares e até 1 ponto pela justificativa e conclusão. Álgebra incompleta, errada, sem notação matemática zera a questão. Passagens algébricas não detalhadas encejarão redução de 0,5 ponto por cada.

```python
import matplotlib.pyplot as plt
import numpy as np

# Função a ser plotada
def f(x):
    return 3*x**3 - 150*np.sin(x)**2 + 2

# Intervalo de valores de x
x = np.linspace(-8, 8, 1000)

# Cálculo dos valores de y
y = f(x)

# Plot da função
plt.plot(x, y, color="blue", linewidth=2)

# Pontos Criticos
P1 = (0, 2)
P2 = (np.pi/2, 2 + 3*(np.pi/2)**2 - 150)

# Marcação dos pontos
plt.scatter(P1[0], P1[1], color="red", label=f"P1 {P1}")
plt.scatter(P2[0], P2[1], color="blue", label=f"P2 {P2}")

# Linha conectando os pontos P1 e P2 diretamente
plt.plot([P1[0], P2[0]], [P1[1], P2[1]], color="purple", linestyle="dotted", label="Tangente P1-P2")

# Estender a linha para além de P2
extension_factor = 1.5
extension_x = P2[0] + (P2[0] - P1[0]) * extension_factor
extension_y = P2[1] + (P2[1] - P1[1]) * extension_factor
plt.plot([P1[0], extension_x], [P1[1], extension_y], color="purple", linestyle="dotted")

# Linhas que conectam P1 e P2 aos eixos x e y
plt.plot([P2[0], P2[0]], [0, P2[1]], color="green", linestyle="--", linewidth=1)  # Linha de P2 ao eixo y
plt.plot([0, P1[0]], [P1[1], P1[1]], color="orange", linestyle="--", linewidth=1)  # Linha de P1 ao eixo x

# Triângulo
plt.fill([P1[0], 0, P2[0]], [P1[1], 0, P2[1]], color="gray", alpha=0.3)

# Eixo x
plt.plot([-8, 8], [0, 0], color="black", linewidth=0.5)

# Eixo y
plt.plot([0, 0], [-300, 300], color="black", linewidth=0.5)

# Título do gráfico
plt.title("Gráfico da função 3x^3 - 150sin(x)^2 + 2")

# Rótulos dos eixos
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Legendas
plt.legend()

# Mostrar o gráfico
plt.show()
```

*Justificacao dos pontos criticos:*

P1:Porque quando o x assume 0 permite observar o comportamento da curva

P2:onde a funcao assume o ponto de inflexao da curva ou seja o meno

*=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=*

*Substituição dos pontos na função:*

Para $x_1$, $x = 0$, encontramos $y_1 = 2$.

Para $x_2$, $x = \dfrac{\pi}{2}$, encontramos $y_2 = -148 + \dfrac{3\pi^3}{8}$.

*Cálculo da diferença dos valores de $y$:*

$\Delta y = y_2 - y_1 = -148 + \dfrac{3\pi^3}{8} - 2 = -150 + \dfrac{3\pi^3}{8}$.

*Cálculo da diferença dos valores de $x$:*

$\Delta x = x_2 - x_1 = \dfrac{\pi}{2} - 0 = \dfrac{\pi}{2}$.

*Cálculo do coeficiente angular:*

$m = \dfrac{\Delta y}{\Delta x} = \dfrac{-150 + \dfrac{3\pi^3}{8}}{\dfrac{\pi}{2}}$.

Simplificando:

$m = 2 \left(-150 + \dfrac{3\pi^3}{8}\right) = 3 \left(-400 + \pi^3\right)$.

Coeficiente angular numérico:

$m \approx {-88.090554202}$.

$$\frac{{(2)^2 + 5(2) + 6}}{((2)^2 - 12(2) + 20} = \frac{4 + 10 + 6}{4 - 24 + 20} = \frac{0}{0}.$$

Obtemos uma forma indeterminada $\frac{0}{0}$, o que significa que precisamos manipular algebricamente a expressão para resolver o limite.

Para simplificar a expressão, vamos fatorar o numerador e o denominador:



- O numerador $x^2 + 5x + 6$ pode ser fatorado como $(x - 2)(x - 3)$.
- O denominador $x^2 - 12x + 20$  pode ser fatorado como $(x - 2)(x + 10)$.

Substituindo as fatorações na expressão do limite, temos:

$$\lim_{x \to -2} \frac{(x - 2)(x - 3)}{(x - 2)(x + 10)}.$$

Assim, a expressão simplificada é:

$$\lim_{x \to -2} \frac{x -3}{x - 10}.$$

Agora, aplicamos a substituição direta novamente:

$$\frac{2 + 3}{2 - 10} = \frac{-1}{-8} = \frac{1}{8}.$$

Portanto, o limite da função original quando $x$ se aproxima de $2$ é $\frac{1}{8}$.

## Questão 3

Usando apenas recursos gráficos e as análises de aproximações laterais. Encontre os limites especificados nos itens $a$, $b$, $c$, $d$, $e$, e $f$ para a função definida por partes $f(t)$ descrita à seguir. Observe que todas as ferramentas (gráficos, tabelas, análises e códigos, que você usar para a sua análise devem estar registradas neste caderno.

$$
f(t) =
\begin{cases}
t^2 & \text{para } t < -3 \\
t + 6 & \text{para } -1 < t < 3 \\
t^2 - t & \text{para } t < 3 \\
3t - 2 & \text{para } t \geq 3
\end{cases}
$$



(a) $$\lim_{{t \to -3}} f(t)$$

(b) $$ \lim_{{t \to -1^+}} f(t)$$

(c) $$ \lim_{{t \to 3^-}} f(t)$$

(d) $$ \lim_{{t \to 0}} f(t)$$

(e) $$ \lim_{{t \to \infty}} f(t)$$

(f) $$ \lim_{{t \to -\infty}} f(t)$$


Valor da Questão: cada item vale até 0,5 pontos. A falta do registro da análise zera o item. A inconsistência no processo de análise também zera o item. A inconsistência será visível quando a análise gráfica divergir das análise por aproximação lateral ou da conclusão apresentada.


```python
# Letra a
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

t = sp.symbols('t')
f = t**2

limite = sp.limit(f, t, -3)

t_values = np.linspace(-3.1, -2.9, 400)
f_values = np.power(t_values, 2)

plt.plot(t_values, f_values, label=r'$f(t) = t^2$')
plt.axvline(x=-3, color='r', linestyle='--', label=r'$t = -3$')
plt.xlabel(r'$t$')
plt.ylabel(r'$f(t)$')
plt.title('Gráfico de $f(t) = t^2$')
plt.legend()
plt.grid(True)
plt.show()

print("O limite de f(t) quando t se aproxima de -3 é:", limite)
```

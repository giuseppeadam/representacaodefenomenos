# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



<h1> Resposta</h1>

### Passo 1: Definir a integral

A quantidade de biomassa acumulada é proporcional à área sob a curva $ y $ no intervalo de 0 a 2. Portanto, precisamos calcular:

$$
\int_{0}^{2} x^5 (2 - x) \, dx
$$

### Passo 2: Expandir a função

Expandir a função dentro da integral para facilitar a integração:

$$
x^5 (2 - x) = 2x^5 - x^6
$$

### Passo 3: Integrar termo a termo

Agora, integramos termo a termo:

$$
\int_{0}^{2} (2x^5 - x^6) \, dx = 2 \int_{0}^{2} x^5 \, dx - \int_{0}^{2} x^6 \, dx
$$

### Passo 4: Calcular as integrais

Calcular as integrais individuais:

$$
\int_{0}^{2} x^5 \, dx = \left. \frac{x^6}{6} \right|_{0}^{2} = \frac{2^6}{6} - \frac{0^6}{6} = \frac{64}{6} = \frac{32}{3}
$$

$$
\int_{0}^{2} x^6 \, dx = \left. \frac{x^7}{7} \right|_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7} = \frac{128}{7}
$$

### Passo 5: Substituir os resultados

Substituir os resultados das integrais na expressão:

$$
2 \int_{0}^{2} x^5 \, dx - \int_{0}^{2} x^6 \, dx = 2 \left( \frac{32}{3} \right) - \frac{128}{7}
$$

Simplificar a expressão:

$$
2 \left( \frac{32}{3} \right) = \frac{64}{3}
$$

Portanto:

$$
\frac{64}{3} - \frac{128}{7}
$$

Para subtrair essas frações, precisamos de um denominador comum. O denominador comum é 21:

$$
\frac{64}{3} = \frac{64 \times 7}{3 \times 7} = \frac{448}{21}
$$

$$
\frac{128}{7} = \frac{128 \times 3}{7 \times 3} = \frac{384}{21}
$$

Agora subtraímos:

$$
\frac{448}{21} - \frac{384}{21} = \frac{64}{21}
$$

### Resultado Final

Portanto, a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos é:

$$
\frac{64}{21} \, \text{unidades de biomassa}
$$

```python
import matplotlib.pyplot as plt
import numpy as np


def crescimento(x):
    return x**5 * (2 - x)


x = np.linspace(0, 2, 1000)
y = crescimento(x)


area = np.trapz(y, x)


plt.figure(figsize=(10, 6))
plt.plot(x, y, color="red", label='Crescimento do cacto')
plt.fill_between(x, y, alpha=0.3,color="pink", label=f'Área sob a curva (Biomassa acumulada): {area:.2f}')
plt.xlabel('Tempo (anos)')
plt.ylabel('Altura do cacto (m)')
plt.title('Crescimento do cacto ao longo do tempo')
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

<h1> Resposta</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


x = sp.symbols('x')
F = x**5 * (2 - x)


area = sp.integrate(F, (x, 0, 2))
area_val = area.evalf()


F_lambdified = sp.lambdify(x, F, 'numpy')


x_vals = np.linspace(0, 2, 1000)
y_vals = F_lambdified(x_vals)


plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, color="red",label='$F(x) = x^5 (2 - x)$')
plt.fill_between(x_vals, y_vals, color="red", alpha=0.3, label=f'Área sob a curva: {area_val:.2f}')
plt.xlabel('x')
plt.ylabel('$F(x)$')
plt.title('Área sob a função $F(x) = x^5 (2 - x)$')
plt.legend()
plt.grid(True)
plt.annotate(f'Área = {area_val:.2f}', xy=(1, F_lambdified(1)), xytext=(1.5, 1),
             )
plt.show()


print(f'A área sob a curva de F(x) entre 0 e 2 é: {area_val:.2f}')
```

 precisamos integrar a aceleração para obter a velocidade e, em seguida, integrar a velocidade para obter a posição.

### Passo 1: Encontrar a velocidade

A aceleração é a derivada da velocidade. Portanto, integramos a aceleração para encontrar a velocidade.

$$
a(t) = \cos(\pi t)
$$

A integral de \( \cos(\pi t) \) é:

$$
v(t) = \int \cos(\pi t) \, dt = \frac{1}{\pi} \sin(\pi t) + C
$$

Usamos a condição inicial \( v(0) = \frac{1}{2\pi} \) para encontrar a constante \( C \).

$$
v(0) = \frac{1}{\pi} \sin(0) + C = \frac{1}{2\pi}
$$

$$
0 + C = \frac{1}{2\pi}
$$

Portanto,

$$
C = \frac{1}{2\pi}
$$

A velocidade é então:

$$
v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi}
$$

### Passo 2: Encontrar a posição

A velocidade é a derivada da posição. Portanto, integramos a velocidade para encontrar a posição.

$$
v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi}
$$

A integral de \( v(t) \) é:

$$
x(t) = \int \left( \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi} \right) dt
$$

Integrando termo a termo:

$$
x(t) = \frac{1}{\pi} \int \sin(\pi t) \, dt + \frac{1}{2\pi} \int 1 \, dt
$$

A integral de \( \sin(\pi t) \) é \( -\frac{1}{\pi} \cos(\pi t) \):

$$
x(t) = \frac{1}{\pi} \left( -\frac{1}{\pi} \cos(\pi t) \right) + \frac{1}{2\pi} t + C
$$

$$
x(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + C
$$

Usamos a condição inicial \( x(0) = 0 \) para encontrar a constante \( C \):

$$
x(0) = -\frac{1}{\pi^2} \cos(0) + \frac{1}{2\pi} \cdot 0 + C = 0
$$

$$
-\frac{1}{\pi^2} + C = 0
$$

$$
C = \frac{1}{\pi^2}
$$

Portanto, a posição é:

$$
x(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + \frac{1}{\pi^2}
$$

### Passo 3: Calcular a distância total percorrida de 0 a 1,5 segundos

A posição em \( t = 1,5 \) segundos é:

$$
x(1.5) = -\frac{1}{\pi^2} \cos(1.5\pi) + \frac{1}{2\pi} \cdot 1.5 + \frac{1}{\pi^2}
$$

Sabendo que \( \cos(1.5\pi) = 0 \):

$$
x(1.5) = 0 + \frac{1.5}{2\pi} + \frac{1}{\pi^2}
$$

$$
x(1.5) = \frac{1.5}{2\pi} + \frac{1}{\pi^2}
$$

Simplificando:

$$
x(1.5) = \frac{0.75}{\pi} + \frac{1}{\pi^2}
$$

Portanto, a distância total percorrida pelo objeto nos primeiros 1,5 segundos é:

$$
\frac{0.75}{\pi} + \frac{1}{\pi^2} \approx 0.2387 + 0.1013 \approx 0.34 \, \text{metros}
$$

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


<h1> Resposta</h1>


### Passo 1: Esboçar a Região

Primeiro, esboçamos a região delimitada pelas funções. As funções são:
- $ y = \sqrt{x} $
- $ y = 3 $
- Eixo $ y $

A interseção de $ y = \sqrt{x} $ e $ y = 3 $ ocorre quando:
$$
3 = \sqrt{x} \Rightarrow x = 9
$$

Portanto, a região de interesse é para $ x $ entre $ 0 $ e $ 9 $, e $ y $ entre $ 0 $ e $ 3 $.

### Passo 2: Expressar $ x $ como função de $ y $

Para usar o método dos discos em torno do eixo $ y $, precisamos expressar $ x $ como uma função de $ y $. A função $ y = \sqrt{x} $ pode ser reescrita como:
$$
x = y^2
$$

### Passo 3: Configurar a Integral

O volume $ V $ do sólido de revolução usando o método dos discos é dado por:
$$
V = \pi \int_{a}^{b} [R(y)]^2 \, dy
$$

Onde $ R(y) $ é a função que descreve o raio do disco, que no nosso caso é $ x = y^2 $. Os limites de integração são de $ y = 0 $ a $ y = 3 $.

Portanto:
$$
V = \pi \int_{0}^{3} (y^2)^2 \, dy
$$

### Passo 4: Integrar

$$
V = \pi \int_{0}^{3} y^4 \, dy
$$

Integramos:
$$
\int_{0}^{3} y^4 \, dy = \left. \frac{y^5}{5} \right|_{0}^{3} = \frac{3^5}{5} - \frac{0^5}{5} = \frac{243}{5}
$$

### Passo 5: Calcular o Volume

Multiplicamos pelo fator $\pi$:
$$
V = \pi \cdot \frac{243}{5} = \frac{243\pi}{5}
$$

Portanto, o volume do sólido obtido pela rotação da região em torno do eixo $ y $ é:
$$
V = \frac{243\pi}{5}
$$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display

# Definindo a função usando sympy
y = sp.symbols('y')
f_inv = y**2

# Calculando os limites de integração
a = 0
b = 3

# Calculando o volume usando o método dos discos
volume = sp.integrate(sp.pi * f_inv**2, (y, a, b))
volume_val = volume.evalf()

# Definindo as funções para plotagem
def f(y):
    return np.sqrt(y)

# Intervalo de y
y_vals = np.linspace(a, b, 400)
x_vals = f(y_vals)

# Gráfico das funções e área de rotação
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$y = \sqrt{x}$')
plt.fill_betweenx(y_vals, 0, x_vals, alpha=0.3, label='Região de rotação')
plt.axhline(y=3, color='r', linestyle='--', label='$y = 3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e eixo $y$')
plt.legend()
plt.grid(True)
plt.show()

# Exibir o valor do volume
print(f'O volume do sólido obtido pela rotação da região em torno do eixo y é: {volume_val:.2f}')
```

### Problema (a)

A função é:
$$
f(x) = 8x - 3 + 5e^{2 - x}
$$

Devemos encontrar a média de $ f(x) $ no intervalo $ [0, 2] $. A fórmula para a média de uma função $ f(x) $ no intervalo $ [a, b] $ é:
$$
\text{Média} = \frac{1}{b - a} \int_{a}^{b} f(x) \, dx
$$

Substituindo $ f(x) $ e os limites de integração:
$$
\text{Média} = \frac{1}{2 - 0} \int_{0}^{2} (8x - 3 + 5e^{2 - x}) \, dx
$$

$$
\text{Média} = \frac{1}{2} \left[ \int_{0}^{2} 8x \, dx - \int_{0}^{2} 3 \, dx + \int_{0}^{2} 5e^{2 - x} \, dx \right]
$$

Calculando cada integral separadamente:

1. $ \int_{0}^{2} 8x \, dx $

$$
\int_{0}^{2} 8x \, dx = 8 \left[ \frac{x^2}{2} \right]_{0}^{2} = 8 \left( \frac{4}{2} - \frac{0}{2} \right) = 8 \times 2 = 16
$$

2. $ \int_{0}^{2} 3 \, dx $

$$
\int_{0}^{2} 3 \, dx = 3 \left[ x \right]_{0}^{2} = 3 (2 - 0) = 6
$$

3. $ \int_{0}^{2} 5e^{2 - x} \, dx $

Para esta integral, fazemos a substituição $ u = 2 - x $.

$$
du = -dx \quad \Rightarrow \quad dx = -du
$$

Os limites de integração mudam de $ x = 0 $ para $ u = 2 $ e de $ x = 2 $ para $ u = 0 $:

$$
\int_{0}^{2} 5e^{2 - x} \, dx = 5 \int_{2}^{0} e^u \, (-du) = 5 \int_{0}^{2} e^u \, du = 5 \left[ e^u \right]_{0}^{2} = 5 (e^2 - e^0) = 5 (e^2 - 1)
$$

Somando os resultados:
$$
\text{Média} = \frac{1}{2} \left( 16 - 6 + 5 (e^2 - 1) \right) = \frac{1}{2} \left( 10 + 5e^2 - 5 \right) = \frac{1}{2} (5 + 5e^2) = \frac{5}{2} (1 + e^2)
$$

Portanto, a média da função no intervalo $ [0, 2] $ é:
$$
\frac{5}{2} (1 + e^2)
$$

### Problema (b)

A função de deslocamento é:
$$
x(t) = t^3 - 6t^2 + 9t + 1
$$

Para encontrar a velocidade média no intervalo de $ t = 1 $ segundo até $ t = 4 $ segundos, usamos a fórmula da velocidade média:

$$
v_{\text{média}} = \frac{x(b) - x(a)}{b - a}
$$

Substituindo $ a = 1 $ e $ b = 4 $:
$$
v_{\text{média}} = \frac{x(4) - x(1)}{4 - 1}
$$

Calculando $ x(4) $ e $ x(1) $:

1. $ x(4) = 4^3 - 6(4^2) + 9(4) + 1 $

$$
x(4) = 64 - 96 + 36 + 1 = 5
$$

2. $ x(1) = 1^3 - 6(1^2) + 9(1) + 1 $

$$
x(1) = 1 - 6 + 9 + 1 = 5
$$

Portanto:
$$
v_{\text{média}} = \frac{5 - 5}{4 - 1} = \frac{0}{3} = 0
$$

Portanto, a velocidade média do objeto no intervalo de $ t = 1 $ segundo até $ t = 4 $ segundos é $ 0 $ metros por segundo.

Nos ítens a, b a seguir, encontre a família de antiderivadas das seguintes funções:

a. $f(x) = 4x^3 - 6x^2 + 2x - 5 $.

b. $f(x) = \frac{1}{x} + e^x$ para todo $x\neq 0$.




<h1> Resposta</h1>


### Item (a)

A função é:
$$
f(x) = 4x^3 - 6x^2 + 2x - 5
$$

Integramos cada termo da função separadamente:

1. $\int 4x^3 \, dx$

$$
\int 4x^3 \, dx = 4 \cdot \frac{x^4}{4} = x^4
$$

2. $\int -6x^2 \, dx$

$$
\int -6x^2 \, dx = -6 \cdot \frac{x^3}{3} = -2x^3
$$

3. $\int 2x \, dx$

$$
\int 2x \, dx = 2 \cdot \frac{x^2}{2} = x^2
$$

4. $\int -5 \, dx$

$$
\int -5 \, dx = -5x
$$

Somando todas as antiderivadas e adicionando a constante de integração $C$:

$$
\int (4x^3 - 6x^2 + 2x - 5) \, dx = x^4 - 2x^3 + x^2 - 5x + C
$$

Portanto, a família de antiderivadas de $f(x)$ é:

$$
x^4 - 2x^3 + x^2 - 5x + C
$$

### Item (b)

A função é:
$$
f(x) = \frac{1}{x} + e^x
$$

Integramos cada termo da função separadamente:

1. $\int \frac{1}{x} \, dx$

$$
\int \frac{1}{x} \, dx = \ln |x|
$$

2. $\int e^x \, dx$

$$
\int e^x \, dx = e^x
$$

Somando todas as antiderivadas e adicionando a constante de integração $C$:

$$
\int \left( \frac{1}{x} + e^x \right) \, dx = \ln |x| + e^x + C
$$

Portanto, a família de antiderivadas de $f(x)$ é:

$$
\ln |x| + e^x + C
$$

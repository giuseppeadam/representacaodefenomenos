<h1> 1:</h1>
<br>
$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} $$$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} $$

podemos usar a seguinte estratégia:

Fatora o numerador:
$$ \frac{\sqrt{x+9} - 3}{x} = \frac{(\sqrt{x+9} - 3)(\sqrt{x+9} + 3)}{x(\sqrt{x+9} + 3)} $$

Simplifica o numerador usando a diferença de quadrados:
$$ \frac{(\sqrt{x+9} - 3)(\sqrt{x+9} + 3)}{x(\sqrt{x+9} + 3)} = \frac{(x+9) - 9}{x(\sqrt{x+9} + 3)} $$
$$ = \frac{x}{x(\sqrt{x+9} + 3)} $$

Cancelar o x em comum:
$$ = \frac{1}{\sqrt{x+9} + 3} $$
Agora resolver a equacao obtida:
$$ \lim_{x \to 0} \frac{1}{\sqrt{x+9} + 3} = \frac{1}{\sqrt{0+9} + 3} $$

$$ = \frac{1}{3+3} $$

$$ = \frac{1}{6} $$

Portanto, o limite é igual a 1/6.

```python
#exercicio1
from sympy import symbols, limit, diff, sin, cos, ln, exp, sqrt, Rational
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = symbols('x')
f = (sqrt(x+9)-3) / (x)

ponto_desejado = 0

def criar_pontos_aproximacao(valor_limite, distancia=[0.1, 0.01, 0.001, 0.0001, 0.00001]):
    pontos_esquerda = [valor_limite - d for d in distancia]
    pontos_direita = [valor_limite + d for d in distancia]
    return pontos_esquerda, pontos_direita

# Reaplicando a função para criar os pontos de aproximação
pontos_esquerda, pontos_direita = criar_pontos_aproximacao(ponto_desejado)

# Recalculando os valores de f(x) para os pontos à esquerda e à direita
valores_esquerda = [f.subs(x, ponto).evalf() for ponto in pontos_esquerda]
valores_direita = [f.subs(x, ponto).evalf() for ponto in pontos_direita]

# Preparando a saída em formato de tabela novamente
tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

display(Math(tabela_latex))
```

Observando o comportamento das aproximacoes laterais, vemos que e limite de $x$ quando tende a 0 e igual a $0,16667$, aproximadamente, correspondendo ao resultado anterior:  $\frac{1}{6}$.

$$\lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} = \lim_{x \to \infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}}$$

$$\lim_{x \to \infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}} = \frac{\lim_{x \to \infty}(25 + \frac{7}{x})}{\lim_{x \to \infty}\sqrt{5 + \frac{2}{x^2}}}$$

$$\lim_{x \to \infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}} = \frac{25 + 0}{\sqrt{5 + 0}}$$

$$\lim_{x \to \infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}} = \frac{25}{\sqrt{5}}$$

<h1> 3: </h1>

$$
\lim_{x \to 1} f(x) \text{ onde } f(x) = \begin{cases}
2x+1 & x \leq 1 \\
x^2 + 1 & x > 1
\end{cases}
$$
como e uma funcao em partes, vamosVerificar se a função é contínua em $ x = 1$
Para que uma função seja contínua em um ponto, seus limites lateral esquerdo e direito nesse ponto devem existir e ser iguais.

Passo 2: Calcular o limite lateral esquerdo

O limite lateral esquerdo é dado por:

$$\lim_{x \to 1^-} f(x) = \lim_{x \to 1^-} (2x+1) = 2(1) + 1 = 3$$

Passo 3: Calcular o limite lateral direito

O limite lateral direito é dado por:

$$\lim_{x \to 1^+} f(x) = \lim_{x \to 1^+} (x^2 + 1) = (1)^2 + 1 = 2$$

Passo 4: Comparar os limites laterais

Como os limites laterais esquerdo e direito não são iguais, a função não é contínua em x = 1. Portanto, o limite da função em x = 1 não existe.

Conclusão:

 o limite desse $\lim_{x \to 1} f(x) $ não existe

<h1> 4: </h1>
<br>
$$\lim_{x \to -\infty} (\sqrt{x^2 - x} - x)$$


Fatora a expressão dentro da raiz quadrada:
$$\lim_{x \to -\infty} (\sqrt{x^2 - x} - x) = \lim_{x \to -\infty} \left(\sqrt{x(x - 1)} - x\right)$$

Divide a expressão por x:
$$\lim_{x \to -\infty} \left(\sqrt{x(x - 1)} - x\right) = \lim_{x \to -\infty} \left(\frac{\sqrt{x(x - 1)}}{x} - \frac{x}{x}\right)$$
Simplifica:
$$\lim_{x \to -\infty} \left(\frac{\sqrt{x(x - 1)}}{x} - \frac{x}{x}\right) = \lim_{x \to -\infty} \left(\frac{\sqrt{x(x - 1)} - x}{x}\right)$$

Usar a regra do limite do quociente:
$$\lim_{x \to -\infty} \left(\frac{\sqrt{x(x - 1)} - x}{x}\right) = \frac{\lim_{x \to -\infty} (\sqrt{x(x - 1)} - x)}{\lim_{x \to -\infty} x}$$

Calcular os limites individuais:
$$\lim_{x \to -\infty} (\sqrt{x(x - 1)} - x) = -\infty$$
$$\lim_{x \to -\infty} x = -\infty$$
Substituir os limites individuais na expressão original:
$$\frac{-\infty}{-\infty} = \infty$$
Portanto, o limite

$$\lim_{x \to -\infty} (\sqrt{x^2 - x} - x) = \infty$$
